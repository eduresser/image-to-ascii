```python
def get_image(url, scale):
    # Carregando a imagem
    response = requests.get(url)
    image = Image.open(BytesIO(response.content))
    image = image.convert("L")
    width, height = image.size
    
    print((width, height), (int(width * scale), int(height * scale)))
    
    return image.resize((int(width * scale), int(height * scale)))
```

```python
def create_ascii_matrix(image_array):
    height, width = image_array.shape
    ascii_matrix = np.zeros((height, width), dtype=int)
    
    colors = set()

    for row in image_array:
        for pixel in row:
            colors.add(pixel)
    
    colors = list(colors)

    for i, row in enumerate(image_array):
        for j, pixel in enumerate(row):
            if pixel < 255 * 1 / 16:
                ascii_matrix[i, j] = 1
            elif pixel < 255 * 4 / 16:
                if (i % 4 == 0 and j % 4 == 0):
                    ascii_matrix[i, j] = 0
                else:
                    ascii_matrix[i, j] = 1
            elif pixel < 255 * 7 / 16:
                if (i % 2 == 0 and j % 2 == 0):
                    ascii_matrix[i, j] = 0
                else:
                    ascii_matrix[i, j] = 1
            elif pixel < 255 * 9 / 16:
                if (i % 2 == 0 and j % 2 == 0) or (i % 2 != 0 and j % 2 != 0):
                    ascii_matrix[i, j] = 1
                else:
                    ascii_matrix[i, j] = 0
            elif pixel < 255 * 12 / 16:
                if (i % 2 == 0 and j % 2 == 0):
                    ascii_matrix[i, j] = 1
                else:
                    ascii_matrix[i, j] = 0
            elif pixel < 255 * 15 /16:
                if (i % 4 == 0 and j % 4 == 0):
                    ascii_matrix[i, j] = 1
                else:
                    ascii_matrix[i, j] = 0
            else:
                ascii_matrix[i, j] = 0

    return ascii_matrix
```

```python
def get_braille_string(matrix):
    rows = 4
    columns = 2
    
    string = ''
    for i in range(0, matrix.shape[0] - rows + 1, rows):
        for j in range(0, matrix.shape[1] - columns + 1, columns):
            block = matrix[i:i+rows, j:j+columns]
            block_code_np = [block[0, 0], block[1, 0], block[2, 0], block[0, 1], block[1, 1], block[2, 1], block[3, 0], block[3, 1]]
            block_code_str = [str(x) for x in block_code_np]
            block_code = ''.join(block_code_str)
            
            string += braille[block_code]
        string += '\n'

    return string
```

```python
def image_to_braille(url, ratio=1):
    image = get_image(url, ratio)
    image_array = np.array(image)
    matrix = 1-create_ascii_matrix(image_array)
    braille_string = get_braille_string(matrix)
    return braille_string
```
